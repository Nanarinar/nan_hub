```python
import numpy as np
from keras.datasets import fashion_mnist
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline


""" Label	Description
    0	T-shirt/top
    1	Trouser
    2	Pullover
    3	Dress
    4	Coat
    5	Sandal
    6	Shirt
    7	Sneaker
    8	Bag
    9	Ankle boot """
```

# Fetch MNIST data

```python
((trainX, trainY), (testX, testY)) = fashion_mnist.load_data()
#print(type(trainX), trainX)
```

```python
print('trainX', type(trainX), trainX.shape)
```

```python
plt.imshow(trainX[101].reshape(28,28), cmap = plt.cm.binary, interpolation = 'nearest'), plt.axis("on")
plt.show()
```

# The distribution of TRAINING data

```python
digit_train, counts_train = np.unique(trainY, return_counts = True)
#find the unique elements matrix and show how many values correspond to each matrix entry
distribution_train = dict(zip(digit_train, counts_train)) #dict to form a "a: number" shape of output; zip takes 
#"a" from digit train and "number" from counts train
print(distribution_train )
```

```python
plt.bar(list(distribution_train.keys()),distribution_train.values(),width =0.6) #keys - LHS of dict result from above
plt.xlabel('Digit Number')
plt.ylabel('Counts')
plt.show()
```

# The distribution of TEST data

```python
digit_test, counts_test = np.unique(testY, return_counts = True) #same as before but for test output data

distribution_test = dict(zip(digit_test, counts_test))
print(distribution_test)
```

```python
plt.bar(list(distribution_test.keys()),distribution_test.values(),width =0.6)
plt.xlabel('Digit Number')
plt.ylabel('Counts')
plt.show()
```

# Inspect some examples of the scaled image

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

trainX = trainX.reshape(60000,784) #transform from 3D array to 2D by changing 28x28 array to 

trainX_unshuffled_scaled = scaler.fit_transform(trainX.astype(np.float64))
```

```python
plt.imshow(trainX[101].reshape(28,28), cmap = plt.cm.binary, interpolation = 'nearest') #plt.axis("off")
plt.imshow(trainX_unshuffled_scaled[101].reshape(28,28), cmap = plt.cm.binary, interpolation = 'nearest')
plt.show()
```

```python
fig_object, ax_object = plt.subplots(1, 10, figsize=(12,5))
ax_object = ax_object.reshape(10,)
      
for i in range(len(ax_object)):
    ax = ax_object[i]
    ax.imshow(trainX[6500*i].reshape(28,28), cmap = plt.cm.binary, interpolation = 'nearest')
    ax.set_title(i)       
plt.show()

fig_object, ax_object = plt.subplots(1, 10, figsize=(12,5))
ax_object = ax_object.reshape(10,)
      
for i in range(len(ax_object)):
    ax = ax_object[i]
    ax.imshow(trainX_unshuffled_scaled[6500*i].reshape(28,28), cmap = plt.cm.binary, interpolation = 'nearest')
    ax.set_title(i)       
plt.show()
```
